```python
import bs4
import requests
from bs4 import BeautifulSoup
res = requests.get("https://www.rottentomatoes.com/") # sends request to website to let me web scrape
```

```python
type(res) # To see if request is valid
```

```python
res.text # Gives you all the html data, but not nicely
```

```python
soup = BeautifulSoup(res.text,"lxml") # Makes html output look nicer
```

```python
soup # to see data nicely formatted
```

```python
soup.title
```

```python
soup.select('title')
```

```python
sell = soup.select('title')[0].getText()
```

```python
soup.select('li') # Brings out all the data with li(list) elements
```

```python
trending = soup.select(".trending-bar__link") #Selects all elements in the '.trending-bar__link' class
```

```python
trending #All elements in class
```

```python
for item in soup.select(".trending-bar__link"): # Have to use loop to go into list and print out text version of trending movies
    print(item.text)
```
