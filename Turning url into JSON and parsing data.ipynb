```python
from urllib.request import urlopen #import
import json

link = "https://jsonplaceholder.typicode.com/todos" #url to be converted into JSON file

response = urlopen(link) #url response

jsonData = json.loads(response.read()) #initialzing data from url into json


for line in jsonData:
    print('title: ' + line['title']) #prints all key values for title
```
